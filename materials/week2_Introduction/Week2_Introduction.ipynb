### 2.1 Basic operations in matrix
Define $A \in \mathbf{R}^{5 \times 6}, B \in \mathbf{R}^{5 \times 6}$


```python
A = torch.randint(0, 5, (5, 6))
print(A)

B = torch.randint(0, 5, (5, 6))
print(B)
```

```python
A_np = np.random.randint(0, 5, (5, 6))
print(A_np)

B_np = np.random.randint(0, 5, (5, 6))
print(B_np)
```

* **Add**: $C = A + B$
* **Substract**: $C = A - B$

```python
# implement A + B in torch
C = A + B
print(C)

# implement A + B in numpy
C_np = A_np + B_np
print(C_np)

# implement A - B in torch
C = A - B
print(C)

# implement A - B in numpy
C_np = A_np - B_np
print(C_np)
```

* **Element-wise Multiplication** $C = A \odot B \in \mathbf{R}^{m \times n}, \forall i,j$, which means $C_{i, j} = A_{i, j} * B_{i, j}$.

```python
# implement A * B in torch
C = A * B
print(C)

# implement A * B in numpy
C_np = A_np * B_np
print(C_np)
```

* **Multiplication** $C = A \times B^T$, which means $C_{i,j} = âˆ‘_{l=1}^{n}A_{i,l} * B_{l, j}$

* $A_{m\times n}, B_{n \times k}$ so we have output $C_{m \times k}$

```python
# implement A * B in torch
C = torch.matmul(A, B.T)
print(C)

# implement A * B in numpy
C_np = np.matmul(A_np, np.transpose(B_np))
print(C_np)
```

```python
# some other ways to implement A * B in torch
C = A @ B.T
print(C)

C = A.mm(B.T)
print(C)
```

## 3 Create dataset and dataloader by Gaussian distribution

```python
import torch
import matplotlib.pyplot as plt

# Set random seed for reproducibility
torch.manual_seed(42)

# Parameters for two Gaussian distributions
num_samples = 500  # Number of samples per class
dim = 2  # Dimensionality of the data (2D for visualization)

# Class 0 (Mean = [-2, -2], Covariance = Identity)
mean_0 = torch.tensor([-2.0, -2.0])
cov_0 = torch.eye(dim) * 1.0  # Identity covariance

# Class 1 (Mean = [2, 2], Covariance = Identity)
mean_1 = torch.tensor([2.0, 2.0])
cov_1 = torch.eye(dim) * 1.0  # Identity covariance

# Sample points from the Gaussian distributions
class_0 = torch.distributions.MultivariateNormal(mean_0, cov_0).sample((num_samples,))
class_1 = torch.distributions.MultivariateNormal(mean_1, cov_1).sample((num_samples,))

# Create labels
labels_0 = torch.zeros(num_samples, dtype=torch.long)  # Label 0
labels_1 = torch.ones(num_samples, dtype=torch.long)   # Label 1

# Combine data
X = torch.cat([class_0, class_1], dim=0)
y = torch.cat([labels_0, labels_1], dim=0)

# Shuffle dataset
indices = torch.randperm(X.size(0))
X, y = X[indices], y[indices]

# Plot the dataset
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k", alpha=0.7)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Toy Binary Classification Dataset (Two Gaussian Distributions)")
plt.show()

```

### 3.1 Create Dataset

```python
from torch.utils.data import Dataset

class GaussianDataset(Dataset):
    def __init__(self, X, y):
        """
        A PyTorch Dataset for the generated Gaussian-distributed classification data.

        Args:
            X (torch.Tensor): Features (2D points).
            y (torch.Tensor): Labels (0 or 1).
        """
        self.X = X
        self.y = y

    def __len__(self):
        """Returns the total number of samples in the dataset."""
        return len(self.X)

    def __getitem__(self, index):
        """Retrieves a single sample from the dataset."""
        return self.X[index], self.y[index]
```

### 3.2 Create Dataloader

```python
from torch.utils.data import DataLoader

# Define batch size
batch_size = 32

# Create the dataset
dataset = GaussianDataset(X, y)

# Create DataLoader
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

### 3.3 Data Iteration Visualization

```python
# Iterate through the dataloader
for batch_idx, batch in enumerate(dataloader):
    data, label = batch
    print("data's shape: ", data.shape)
    print("label's shape: ", label.shape)
    
    print("The first data in the batch: ", data[0])
    print("The first label in the batch: ", label[0])
    break
```

```python
# Visualize each batch
for batch_idx, batch in enumerate(dataloader):

    data, label = batch
    plt.figure(figsize=(6, 6))
    plt.scatter(data[:, 0], data[:, 1], c=label, cmap="coolwarm", edgecolors="k", alpha=0.7)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title(f"Batch {batch_idx+1}")
    plt.show()

    # Break after a few batches for visualization
    if batch_idx == 2:
        break
```
